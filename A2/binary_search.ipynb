```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    comparisons = 0

    while left <= right:
        mid = (left + right) // 2

        # First comparison: arr[mid] == target
        comparisons += 1
        if arr[mid] == target:
            return comparisons  # Target found

        # Second comparison: arr[mid] < target
        comparisons += 1
        if arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return comparisons  # Target not found after all comparisons
```

```python
def estimate_average_cost(array_sizes, trials_per_size):
    average_costs = []
    
    for n in array_sizes:
        total_comparisons = 0
        # Generate a sorted array of random integers
        arr = sorted(random.sample(range(1, n ** 3), n))
        
        for _ in range(trials_per_size):
            target = random.randint(1, n ** 3)
            comparisons = binary_search(arr, target)
            total_comparisons += comparisons
        
        average_comparisons = total_comparisons / trials_per_size
        average_costs.append(average_comparisons)
        print(f"Array Size: {n}, Average Comparisons: {average_comparisons:.2f}")
    
    return average_costs
```

```python
# Parameters
array_sizes = [2000,4000,6000,8000,10000,12000,14000,16000,18000,20000]
trials_per_size = 100
```

```python
average_costs = estimate_average_cost(array_sizes, trials_per_size)

# Plotting the results
plt.figure(figsize=(12, 7))
plt.plot(array_sizes, average_costs, marker='o', linestyle='-')
plt.title('Average Cost of Binary Search vs. Array Size', fontsize=16)
plt.xlabel('Array Size', fontsize=14)
plt.ylabel('Average Number of Comparisons', fontsize=14)
plt.grid(True)

plt.xticks(array_sizes, [str(size) for size in array_sizes], fontsize=12)
plt.yticks(fontsize=12)


# Show the plot
plt.tight_layout()
plt.show()
```
